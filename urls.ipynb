```python
import pandas as pd
df_main= pd.read_csv('weights_Catalog Data Set - Sheet1.csv')
df_main
```

```python
df = pd.read_csv('image_indexes_v2.csv')
df
```

```python
# Assuming df_main is a DataFrame containing mapping between Product_id and Product_Name

# Create a dictionary mapping product IDs to product names for faster lookup
product_id_to_name = df_main.set_index('Product_id')['Product_Name'].to_dict()

# Extract product IDs from image names
df['product_id'] = df['image name'].str.split('_').str[0]

# Map product IDs to product names using the dictionary
df['Product_Name'] = df['product_id'].map(product_id_to_name)
```

```python
# df.to_csv('urls.csv', index=False)
df
```

```python
df['image name'] = df['image name'].str.replace('.jpg', '')
```

```python
df2 = pd.read_csv('weights_Catalog Data Set - Sheet1.csv')
df2
```

```python
# Remove duplicates from df2 based on 'Downloaded_Image_Id'
df2_unique = df2.drop_duplicates(subset='Downloaded_Image_Id')

# Merge only the 'Url' column from df2 to df1 based on 'Downloaded_Image_Id'
merged_df = pd.merge(df, df2_unique[['Downloaded_Image_Id', 'Image_Url']], left_on='image name', right_on='Downloaded_Image_Id', how='left')
merged_df.drop('Downloaded_Image_Id', axis=1, inplace=True)
merged_df
```

```python
# Check if the indexes of 'image name' in df are the same as 'image name' in df2
are_indexes_equal = (df['image name'].index == merged_df['image name'].index).all()

# Display the result
print(f"The indexes of 'image name' in df are {'equal' if are_indexes_equal else 'not equal'} to the indexes of 'image name' in df2.")
```
